# Quantum Teleportation

```python
q = QuantumRegister(2,'q')
c = ClassicalRegister(2,'c')
qc = QuantumCircuit(q,c)
qc.i(1)
qc.cx(1,0)
qc.barrier()
qc.h(1)
qc.measure([1],[1])
qc.barrier()
#qc.z(0)
#qc.cx(1,0)
qc.measure([0],[0])
qc.barrier()
qc.draw('mpl')
```

```python
job = execute(qc, backend = Aer.get_backend('qasm_simulator'), shots = 1024)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```
